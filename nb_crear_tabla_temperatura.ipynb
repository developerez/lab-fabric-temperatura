```python
from pyspark.sql.functions import year

# 1. Leer todos los archivos CSV dentro de la carpeta
df = spark.read.csv("Files/temperatura_dmc_flat", header=True, inferSchema=True)

# 2. Guardarlos en formato Delta en una carpeta de tablas
df.write.format("delta").mode("overwrite").save("Tables/temperatura_dmc")

# 3. Registrar la tabla en Spark SQL (solo si no existe), usando la misma ubicaci√≥n
spark.sql("""
CREATE TABLE IF NOT EXISTS temperatura_dmc
USING DELTA
LOCATION 'Tables/temperatura_dmc'
""")

```

```python
spark.table("temperatura_dmc").withColumn("anio", year("time")).write.option("mergeSchema", "true").format("delta").mode("overwrite").saveAsTable("temperatura_dmc")

```
