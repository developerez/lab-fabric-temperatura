```python
from notebookutils import fs

source_path = "Files/Ministerio de Ciencias/Datos-CambioClimatico/refs/heads/main/output/temperatura_dmc"
target_path = "Files/temperatura_dmc"

# Crear carpeta destino si no existe
fs.mkdirs(target_path)

# Contador de archivos movidos
count = 0

for year_folder in fs.ls(source_path):
    print(f"ðŸ“ Carpeta: {year_folder.path}")
    if year_folder.isDir:
        files = fs.ls(year_folder.path)
        for file in files:
            print(f"ðŸ”Ž Revisando: {file.name}")
            if file.name.endswith(".csv"):
                origin = file.path
                dest = f"{target_path}/{file.name}"
                try:
                    fs.mv(origin, dest, True)
                    print(f"âœ… Movido: {origin} â†’ {dest}")
                    count += 1
                except Exception as e:
                    print(f"âš ï¸ Error al mover: {origin} â†’ {dest} â€” {str(e)}")

print(f"ðŸ“Š Total de archivos movidos: {count}")
```
