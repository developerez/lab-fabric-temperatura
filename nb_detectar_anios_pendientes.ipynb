```python
import requests
import re
import json
from notebookutils import mssparkutils  # necesario para Fabric

# --- CONFIGURACI칍N ---
url_api = "https://api.github.com/repos/MinCiencia/Datos-CambioClimatico/contents/output/temperatura_dmc"
ruta_lakehouse = "Files/temperatura_dmc_raw/"

# --- 1. A침os desde GitHub ---
res = requests.get(url_api)
data = res.json()
annios_github = sorted([
    int(item["name"]) for item in data
    if item["type"] == "dir" and re.fullmatch(r"\d{4}", item["name"])
])

# --- 2. A침os ya copiados ---
try:
    df_existentes = spark.read.format("binaryFile").load(ruta_lakehouse + "*/*.csv")
    rutas = df_existentes.select("path").rdd.flatMap(lambda x: x).collect()
    annos_locales = set(int(re.findall(r"temperatura_dmc_raw/(\d{4})/", path)[0]) for path in rutas)
except:
    annos_locales = set()

# --- 3. A침os nuevos ---
anios_nuevos = sorted(set(annios_github) - annos_locales)

# --- 4. Mostrar para validaci칩n (opcional)
print("游녤 A침os detectados en GitHub:", annios_github)
print("游녤 A침os ya presentes en Lakehouse:", sorted(annos_locales))
print("游녤 A침os nuevos pendientes por copiar:", anios_nuevos)

# --- 5. Devolver como salida oficial para ForEach
mssparkutils.notebook.exit(json.dumps(anios_nuevos))
```
