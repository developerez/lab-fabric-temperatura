```python
import pandas as pd
import requests
import re

# 1. Llamar al API de GitHub para listar carpetas por año
api_url = "https://api.github.com/repos/MinCiencia/Datos-CambioClimatico/contents/output/temperatura_dmc"
res = requests.get(api_url)
data = res.json()

# 2. Extraer nombres de carpetas tipo año
annios = sorted([
    int(item["name"]) for item in data
    if item["type"] == "dir" and re.fullmatch(r"\d{4}", item["name"])
])

# 3. Leer tabla existente si ya fue creada
try:
    df_actual = spark.read.table("tabla_annios")
    anios_actuales = set(row["anio"] for row in df_actual.collect())
except:
    anios_actuales = set()

# 4. Comparar y detectar nuevos años
anios_detectados = set(annios)
anios_nuevos = anios_detectados - anios_actuales

# 5. Append si hay nuevos
if anios_nuevos:
    df_nuevos = spark.createDataFrame([(a,) for a in sorted(anios_nuevos)], ["anio"])
    df_nuevos.write.format("delta").mode("append").saveAsTable("tabla_annios")
    print(f"Años nuevos agregados: {sorted(anios_nuevos)}")
else:
    print("No hay años nuevos que agregar.")


```
